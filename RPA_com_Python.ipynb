# OBJETIVO: Criar uma automação para pesquisar Cotações Eletrônicas no site gov.br/compras


## O site www.gov.br/compras contém todos os sistemas utilizados pelo Governo Federal para realizar compras e contratações públicas.

### As Atas de Registros de preços são geradas a partir de pregões ou concorrências, ou através de processos específicos de dispensa de licitação, criando um compromisso do fornecedor que as assinam de entregar os produtos ou serviços no preço registrado, durante o seu período de vigência.

* Ao acessarmos o endereço www.gov.br/compras, será apresentada a seguinte tela:


1. O primeiro passo é clicar no "x" da mensagem "Para começar", para acessar a página principal.

2. Depois é necessário clicar no botão "Cotações Eletrônicas Abertas para Proposta" que fica no final da página:


3. Vai aparecer uma tela com todas as cotações em andamento:


4. Vamos supor que eu seja um fornecedor de Papel A4 e esteja pesquisando procedimentos de compra por dispensa de licitação. Posso pesquisar nesta tela, usando as teclas 'Ctrl + F' e digitando a palavra 'papel', depois pressionando Enter.

5. Caso seja encontrada alguma ocorrência na página, usar a tecla PrScr (Print Screen) para tirar uma foto da tela.
6. Enviar esta foto por e-mail para o endereço do departamento de vendas de papel.
